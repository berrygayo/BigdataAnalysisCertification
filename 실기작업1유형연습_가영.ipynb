### 3. income.csv 파일에서  수입을 5개의 구간으로 나누어 '수입레벨'이라는(상,중상,중,중하,하) 새로운 변수를 생성하시오

```python
L1, L2, L3, L4, L5 = income['income_total'].quantile([0.2,0.4,0.6,0.8,1])
Level=[]
for a in income['income_total']:
    if a<L1:
        Level.append('하')
    elif a<L2:
        Level.append('중하')
    elif a<L3:
        Level.append('중')
    elif a<L4:
        Level.append('중상')
    else:
        Level.append('상')
income['수입레벨']=Level
income
```

### 14.  백화점 데이터(store.csv)에서 최대구매액의 이상치(사분위수 이용)를 중앙값으로 치환하세요.

```python
df = pd.read_csv('./실기작업1유형/store.csv', encoding = 'cp949')
Q1, Q2, Q3 = df['최대구매액'].quantile([0.25,0.5,0.75])
range = Q3 - Q1 

L = Q1 - 1.5 * range 
U = Q3 + 1.5 * range 
print(L,U)
df['최대구매액']= df['최대구매액'].apply(lambda x : Q2 if (x<L) or (x>U) else x)
#df.query(f'최대구매액<{L}')
```

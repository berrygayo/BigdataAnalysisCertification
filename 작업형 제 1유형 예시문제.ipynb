Q1. mtcars 데이터셋(mtcars.csv)의 qsec 컬럼을 최소 최대 척도(min-max scale)로 변환한 후 0.5보다 큰 값을 가지는 레코드 수를 구하시오.

```python
# 직접 코딩 
min = df['qsec'].min()
max = df['qsec'].max()

df['qsec_minmax'] = (df['qsec'] - min)/(max-min)
len(df.query('qsec_minmax>0.5 ')['qsec_minmax'])
```

Q2. mtcars 데이터셋(mtcars.csv)의 qsec 컬럼을 표준정규분포 데이터 표준화 (standardization) 변환 후 최대, 최소값을 각각 구하시오.

```python
m = df['qsec'].mean()
s = df['qsec'].std()
res = (df['qsec']-m)/s 
res.values.min(), res.values.max()
```

```python
Q1 = df['wt'].quantile(0.25)
Q3 = df['wt'].quantile(0.75)
range = Q3 - Q1 
Lower = Q1 - 1.5 * range 
Upper = Q3 + 1.5 * range 
# (Lower,Upper)

outlier = [] 

outlier=(df.query(f'wt<{Lower} or wt>{Upper} ')['wt'].values)
outlier
```
